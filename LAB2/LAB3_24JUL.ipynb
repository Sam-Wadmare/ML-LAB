```python
# Add this to a new cell

# Example: Bar graph of counts for each PRODUCTLINE
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='PRODUCTLINE')
plt.title('Count of Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar graph: Total sales per COUNTRY
plt.figure(figsize=(12,7))
country_sales = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)
sns.barplot(x=country_sales.index, y=country_sales.values)
plt.title('Total Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
